```python
books = pd.read_csv('./Books.csv', encoding='latin-1')
users = pd.read_csv('./Users.csv', encoding='latin-1')
ratings = pd.read_csv('./Ratings.csv', encoding='latin-1')
```

```python
x = ratings['user_id'].value_counts() > 50
x[x].shape
```

```python
y= x[x].index
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
ratings_with_books = ratings.merge(books, on='isbn')
```

```python
ratings_with_books.head()
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'rating':'num_of_rating'},inplace=True)
```

```python
number_rating.head()
```

```python
final_rating = ratings_with_books.merge(number_rating, on='title')
```

```python
final_rating.head()
```

```python
final_rating = final_rating[final_rating['num_of_rating'] >= 50]
```

```python
final_rating.head()
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values= 'rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[123,:].values.reshape(1,-1), n_neighbors=10 )
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=10 )
    
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggested books are: \n")
                else:
                    print(j)
```

```python
recommend_book("1984")
```

```python
recommend_book("The Hitchhiker's Guide to the Galaxy")
```

```python
recommend_book("Zoya")
```
